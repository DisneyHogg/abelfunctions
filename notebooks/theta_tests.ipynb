```python
import abelfunctions.riemanntheta.riemann_theta as rt

def finite_sum(g, X, Yinv, T, S, z):
    X = X.astype(numpy.double)
    Yinv = Yinv.astype(numpy.double)
    T = T.astype(numpy.double)
    S = S.astype(numpy.double)
    
    pi = numpy.pi
    z = numpy.reshape(z, (g,1))
    
    x = z.real.astype(numpy.double)
    y = z.imag.astype(numpy.double)
    
    yshift = numpy.dot(Yinv,y)
    yshiftint = numpy.floor(yshift)
    yshiftfrac = yshift - yshiftint
    
    # compute the first factor in the finite sum from Theorem 2
    nshifts = S - yshiftint
    N = S.shape[1]
    factor_one = numpy.zeros((1,N), dtype=numpy.complex)   # stores each first factor
    for k in range(N):
        nshift = nshifts[:,k]
        exponent = 0.5*numpy.dot(nshift.T, numpy.dot(X,nshift)) + numpy.dot(nshift.T,x)
        exponent = 2.j*pi*numpy.complex(exponent)
        factor_one[0,k] = numpy.exp(exponent)
        
    # compute the second factor
    nshiftfracs = S + yshiftfrac
    factor_two = numpy.zeros((1,N), dtype=numpy.double)
    for k in range(N):
        nshiftfrac = nshiftfracs[:,k]
        vn = numpy.sqrt(pi)*numpy.dot(T,nshiftfrac)
        factor_two[0,k] = numpy.exp(-numpy.linalg.norm(vn)**2)
            
    # multiply them together and compute the sum
    terms = factor_one*factor_two
    osc_part = terms.sum()    
    return osc_part

def RiemannTheta(z, Omega, epsilon=1e-8):
    # obtain all the relevant data
    X = Omega.real
    Y = Omega.imag
    Yinv = numpy.linalg.inv(Y)
    T = numpy.linalg.cholesky(Y)
    A = rt.lattice_reduce(T)

    # compute radius and a gross estimation of the integer points
    genus = Omega.shape[0]
    radius = rt.radius(epsilon, genus, T)
    N = numpy.floor(radius)
    nrange = numpy.arange(-N,N+1)
    args = [nrange]*g
    S = numpy.array(list(numpy.broadcast(*numpy.ix_(*args)))).T
    
    # compute the finite sum part
    v = finite_sum(g, X, Yinv, T, S, z)
    
    # compute the exponential part
    y = numpy.reshape(z,(g,1)).imag
    u = numpy.exp(numpy.pi*numpy.dot(y.T, numpy.dot(Yinv,y))).item(0,0)
    return (u,v)
    
```
